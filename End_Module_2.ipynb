Q.5) Create class 

```python
#Que_5
class Math:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def Add(self):
    return self.a+self.b
  def Sub(self):
    return self.a-self.b
  def Mul(self):
    return self.a*self.b
  def Div(self):
    return self.a/self.b
```

```python
x = Math(20,5)
```

```python
print(x.Add())
print(x.Sub())
print(x.Mul())
print(x.Div())
```

Q.6) Read file & count numbers of digit
alphabets & symbols
(hint one can compare using ‘a’ < digit <
‘z‘

```python
#Que_6
a = input("enter the string: ")
d=l=0
for i in a:
  if i.isdigit():
    d+=1
  elif i.isalpha():
    l+=1
  else:
    pass
  print("Letters:",l)
  print("digit:",d)
```

Q.10) Create menu driven code for
1) Accept 2 numbers
2) Add
3) Sub
4) Mul
5) Div

```python
#QUE_10
n1=float(input("Enter the first no.: "))
n2=float(input("Enter the Second no.: "))
print("select operation : ")
op = input("Enter the specific opearation +,-,*,/ : ")
result=0
if op == '+':
  result=n1+n2
elif op == '-':
  result=n1-n2
elif op == '*':
  result=n1*n2
elif op == '/':
  result=n1/n2
else:
  print("invalid")

print(n1,op,n2, ':',result)
```
